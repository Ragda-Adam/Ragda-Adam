# Handle Outliers

```python
# Boxplot for Numerical features before handling outliers

plt.figure(figsize=(16, 10))
sns.boxplot(data=student_data)
plt.ylabel('Features')
plt.xlabel('Value')
plt.title('Boxplot for Numerical Features')

plt.show()
```

```python
# Handle Outliers and replace the values with the median

# Identify numerical columns
numerical_columns = student_data.select_dtypes(include=['int64', 'float64']).columns

#loop over numerical columns to handle outliers
#finding the first quartile and the third quartile to get the interquartile range
#then check for values that are out of range and replace them with median
for col in numerical_columns:

    Q1 = student_data[col].quantile(0.25)
    Q3 = student_data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    median = student_data[col].median()
    student_data[col] = np.where(student_data[col] < lower_bound, median, student_data[col])
    student_data[col] = np.where(student_data[col] > upper_bound, median, student_data[col])
```

```python
# Boxplot for Numerical features after handling outliers

plt.figure(figsize=(18, 12))
sns.boxplot(data=student_data)
plt.ylabel('Features')
plt.xlabel('Value')
plt.title('Boxplot for Numerical Features')

plt.show()
```

# Encoding

```python
# Identify categorical and numerical columns

categorical_columns = student_data.select_dtypes(include=['object','category']).columns
numerical_columns = student_data.select_dtypes(include=['int64', 'float64']).columns
#print("categorical_columns" ,categorical_columns)
#print("numerical_columns", numerical_columns)
#student_data.head()
```

```python
# Exclude G1,G2,G3 from the numerical columns in order not to be scaled

numerical_columns = numerical_columns.drop([ 'G1', 'G2','G3'])

# Scale numerical features
scaler = StandardScaler()
scaled_numerical_data = scaler.fit_transform(student_data[numerical_columns])
```

```python
# One-hot encode categorical features
onehot_encoder = OneHotEncoder(drop='first', sparse=False)
encoded_categorical_columns = onehot_encoder.fit_transform(student_data[categorical_columns])

#  Combine encoded categorical and scaled numerical features
encoded_columns = onehot_encoder.get_feature_names_out(categorical_columns)
all_columns = np.concatenate([encoded_columns, numerical_columns])
```

```python
# Create DataFrame with encoded categorical features and scaled numerical features
data_processed = pd.DataFrame(np.hstack((encoded_categorical_columns, scaled_numerical_data)), columns=all_columns)
```

```python
# Add back G1, G2, G3 to the data_processed DataFrame
data_processed['G1'] = student_data['G1'].values
data_processed['G2'] = student_data['G2'].values
data_processed['G3'] = student_data['G3'].values


# Set float format to 2 decimal places
pd.options.display.float_format = '{:.2f}'.format

# Print the processed DataFrame
print(data_processed.head())
print(data_processed.describe())
```

# Split and Test

```python
#Drop G1, G2, G3 and Grade_Mean to split data

Y= data_processed["Grade_Mean"]
data_processed.drop(columns = ["G1", "G2", "G3", "Grade_Mean"], inplace = True)
data_processed.head()
```

```python
# split the data to train and test sets

X= data_processed
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

# Random Forest Model

```python
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, Y_train)
Y_pred = rf_model.predict(X_test)
```

```python
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)
mae = mean_absolute_error(Y_test, Y_pred)
print ('Mean Squared Error',mse)
print('mean absolute error',mae)
print('R2 Score',r2)
```

```python
# Hyperparameter tuning for RandomForest
param_grid = {
    'n_estimators': [40, 60, 80],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 5, 10],
   'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(estimator=RandomForestRegressor(), param_grid=param_grid, cv=5, scoring='r2', n_jobs=-1)
grid_search.fit(X_train, Y_train)

#Best parameters and model evaluation
best_rf_model = grid_search.best_estimator_
Y_pred = best_rf_model.predict(X_test)
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

#Store the results
results = {}
results['Random Forest (Tuned)'] = {'MSE': mse, 'R2': r2}
print(f"Random Forest (Tuned) - MSE: {mse}, R2: {r2}")
```

```python
# Cross-validation with the best model


cv_scores = cross_val_score(best_rf_model, data_processed, Y, cv=5, scoring='r2') # Pass both features (X) and target (Y)
print('Cross-Validation R2 Scores:', cv_scores)
print('Mean Cross-Validation R2 Score:', np.mean(cv_scores))

# Store the results
results = {
    'Random Forest (Tuned)': {'MSE': mse, 'MAE': mae, 'R2': r2, 'CV_R2': cv_scores, 'Mean_CV_R2': np.mean(cv_scores)} }
print(f"Random Forest (Tuned) - MSE: {mse}, MAE: {mae}, R2: {r2}, Mean CV R2: {np.mean(cv_scores)}")
```
